```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('cases_time_series.csv')
```

```python
pd.set_option('display.max_row', len(data['date']))
pd.set_option('display.max_column', 7)
data
```

```python
daily_data = data.iloc[::3]
```

### Confirmed Cases in India the graph show exponential nature

```python
plt.figure(figsize=(20,10))
plt.style.use('fivethirtyeight')
plt.plot(daily_data['date'], daily_data['totalconfirmed'], label = 'Confirmed Cases', c = 'orange')
plt.title('Active Cases in India', c = 'black')
plt.tick_params(axis='both', labelsize=13)
plt.ylabel('No Of Cases', fontsize=18)
plt.xticks(rotation=80)
plt.legend(loc="upper left", fontsize=25)
plt.savefig('Active_Cases_in_India.jpg')
plt.tight_layout()
```

```python
plt.figure(figsize=(20,10))
plt.style.use('fivethirtyeight')
plt.plot(daily_data['date'], daily_data['totaldeceased'], label = 'No of Death Cases', c = 'red')
plt.title('Deaths Cases in India', c = 'black')
plt.tick_params(axis='both', labelsize=13)
plt.ylabel('Death Cases', fontsize=18)
plt.xticks(rotation=80)
plt.legend(loc="upper left", fontsize=25)
plt.savefig('Deaths_Cases_in_India.jpg')
plt.tight_layout()
```

```python
plt.figure(figsize=(20,10))
plt.style.use('fivethirtyeight')
plt.plot(daily_data['date'], daily_data['totalrecovered'], label = 'No Of Peoples Recovered', c = 'orange')
plt.title('Recovered Cases in India', c = 'black')
plt.tick_params(axis='both', labelsize=13)
plt.ylabel('Recover Cases', fontsize=18)
plt.xticks(rotation=80)
plt.legend(loc="upper left", fontsize=25)
plt.savefig('Recovered_Cases_in_India.jpg')
plt.tight_layout()
```

## Confirmed Vs Active Case in India
#### Both graphs show exponential nature but we can't say that the graph of recovered cases is proper exponential

```python
plt.figure(figsize=(20,10))
plt.style.use('fivethirtyeight')
plt.plot(daily_data['date'], daily_data['totalconfirmed'], label = 'Confirmed Cases', c = 'red', marker = 'o')
plt.plot(daily_data['date'], daily_data['totalrecovered'], label = 'Recovered Cases', c = 'green', marker = 'X')
plt.title('Confirmed VS Recovered India', c = 'black')
plt.tick_params(axis='both', labelsize=13)
plt.ylabel('Num Cases', fontsize=18)
plt.xticks(rotation=80)
plt.legend(loc="upper left", fontsize=25)
plt.savefig('Confirmed_VS_Recovered_India.jpg')
plt.tight_layout()
```

```python
stdata = pd.read_csv('statewise2.csv')
stdata['recovery_rate'] = (stdata['recovered']/stdata['confirmed'])*100
stdata['deaths_rate'] = (stdata['deaths']/stdata['confirmed'])*100
```

```python
pd.set_option('display.max_row', len(stdata['active']))
pd.set_option('display.max_column', 13)
stdata
```

```python
stdata.drop(index = 0, inplace = True)
```

```python
plt.figure(figsize=(20,10))
plt.style.use('fivethirtyeight')
stdata = stdata.sort_values('recovery_rate', ascending=False)
plt.bar(stdata['state'], stdata['recovery_rate'], label = 'Recovery Rate', color = 'green')
plt.title('Statewise Recovery Rate India', c = 'black')
plt.tick_params(axis='both', labelsize=13)
plt.ylabel('Recovery Rate(%)', fontsize=18)
plt.xticks(rotation=90)
plt.legend(loc="upper right", fontsize=25)
plt.savefig('Statewise_Recovery_Rate_India.jpg')
plt.tight_layout()
```

```python
states_daily = pd.read_csv('states_daily.csv')
```

```python
pd.set_option('display.max_row', len(states_daily['date']))
pd.set_option('display.max_column', 40)
states_daily
```

```python
states_daily['date'] = pd.to_datetime(states_daily['date'], format = '%d/%m/%Y')
states_daily
```

```python
flt = states_daily['status'] == 'Confirmed'
states_daily_conf = states_daily[flt]
states_daily_conf
```

```python
states_daily_conf_gp = states_daily_conf.groupby(states_daily_conf['date'].dt.week)
```

```python
max_id = states_daily_conf_gp['Total'].idxmax()
max_val = states_daily_conf.loc[max_id]
max_val['date'].dt.weekday, max_val['date'], max_val['Total']
```

## <b> Weekwise Max Confirmed cases
<p> <b> Friday, Saturday and sunday is the most dangerous day because these days numbers of confirmed cases is more then other days 

```python
labels = []
days = {0 : 'Mon', 1 : 'Tue', 2 : 'Wed', 3 : 'Thu', 4 : 'Fri', 5 : 'Sat', 6 : 'Sun'}
for dy in max_val['date'].dt.weekday:
    labels.append(days[dy])
plt.figure(figsize=(20,10))
plt.style.use('fivethirtyeight')
x = np.arange(len(max_val['Total']))
plt.bar(x, max_val['Total'], color = 'orange', label = 'Confirmed Cases')
plt.xticks(x,labels)
plt.title('Weekwise Max Confirmed Cases')
plt.legend(loc="upper left", fontsize=25)
plt.savefig('Weekwise_Confirmed_Cases.jpg')
plt.tight_layout()
```

```python
states_daily_conf_wk_day = states_daily_conf.groupby(states_daily_conf['date'].dt.weekday)
```

```python
Total = states_daily_conf_wk_day.agg({'Total' :  'sum'})
Total
```

## <b> Weekdaywise Cases:
<p> <b> On wednesday confirmed, deaths and recovered cases is least as compared to other days 

```python
x = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
plt.figure(figsize=(20,10))
plt.style.use('fivethirtyeight')
plt.bar(x, Total['Total'], color = 'orange', label = 'Confirmed Cases')
plt.title('Weekdaywise Confirmed Cases')
plt.legend(loc="upper left", fontsize=25)
plt.savefig('Weekdaywise_Confirmed_Cases.jpg')
plt.tight_layout()
```
