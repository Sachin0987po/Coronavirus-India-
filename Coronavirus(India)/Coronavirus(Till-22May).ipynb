```python
data = pd.read_csv('cases_time_series.csv')
data.head()
```

```python
data['Date'] = pd.to_datetime(data['Date'], format = '%d/%m/%Y') #Change str to panda datetime
data['Total Active'] = data['Total Confirmed'] - data['Total Recovered'] - data['Total Deceased']
```

```python
data.iplot( x = 'Date', y = ['Total Confirmed','Total Active' ,'Total Deceased', 'Total Recovered'], width=4)
```

```python
data.iplot(x = 'Date', y = ['Daily Confirmed', 'Daily Deceased', 'Daily Recovered'], subplots=True, width=3)
```

```python
data.iplot(kind = 'bar', x = 'Date', y = ['Daily Confirmed', 'Daily Deceased', 'Daily Recovered'], subplots=True)
```

```python
test_data = pd.read_csv('tested.csv')
```

```python
test_data['totalsamplestested'] = test_data['totalsamplestested'].interpolate() 
test_data.rename(columns = {'updatetimestamp' : 'Date'}, inplace = True)
test_data.head()
```

```python
test_data['Date'] = pd.to_datetime(test_data['Date'], format = '%d/%m/%Y')
```

```python
df = data.merge(test_data[['totalsamplestested', 'Date']])
df.rename(columns = {'totalsamplestested' : 'Total Testing'} , inplace = True)
df.head()
```

```python
df.iplot(x = 'Date', y = ['Total Confirmed','Total Testing'], width=3, title='Testing VS Confirmed Cases')
```

```python
state_data = pd.read_csv('statewise2.csv')
state_data.drop(index = 0, inplace = True)
state_data['recovery rate'] = (state_data['recovered']/state_data['confirmed'])*100
state_data['deaths rate'] = (state_data['deaths']/state_data['confirmed'])*100
state_data.head()
```

```python
df = state_data[state_data['confirmed'] >= 200]
df.iplot(theme='pearl', kind = 'bar', x = 'state', y = ['confirmed', 'recovered', 'deaths'], title = 'Statewise Cases India', colors=['#ff9933','#90EE90', '#FF6347'])
```

```python
df.iplot(kind = 'bar', x = 'state', y = ['recovery rate', 'deaths rate'], title='Recovery Rate VS Death Rate', colors=['#90EE90', '#FF6347'])
```
